## Activation function / Transfer Function


__1. Sigmoid/logistic Sigmoid Function (0 to 1)__
- __used for binary classification__

```python
#sigmoid is predicting probablities

import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
print(sigmoid(-100))
print(sigmoid(-90))
print(sigmoid(-10))
print(sigmoid(-6))
print(sigmoid(-1))
print(sigmoid(0))
print(sigmoid(1))
print(sigmoid(2))
print(sigmoid(3))
print(sigmoid(90))
print(sigmoid(100))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate array of 1-- values between -7 to +7 using numpy's linespace function
x_values = np.linspace(-7, 7, 100)


# Calculate corresponding y values using the sigmoid function
y_values = sigmoid(x_values)

print(list(zip(x_values, y_values)))

# Plot the sigmoid curve
plt.figure(figsize=(8, 6)) #width, height in inches
plt.plot(x_values, y_values, label='Sigmoid Curve')
plt.title('Sigmoid Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.axhline(0, color='black',linewidth=0.5)    #used to hilight horizontal axis...#value of y
plt.axvline(0, color='black',linewidth=0.5)    #used to hilight vertical axis ..#value of x
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)  #grid with -- texture
plt.legend()
plt.show()
```

__2. Linear Activatoin funstion__

- It simply outputs the input as it is
- y = constant*x....constant is 1



```python
def linear_activation(x, c=1):   #if user didn't mention constat's value then set it by default as 1
    return c * x
linear_activation(3, 3)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def linear_activation(x, c=1):   #if user didn't mention constat's value then set it by default as 1
    return c * x

# Generate array of 100 values range from -5 to +5 using np.linespace
x_values = np.linspace(-5, 5, 100)

# Calculate corresponding y values using the linear activation function
y_values = linear_activation(x_values, c=2)  # x is 2 then y will 4

# Plot the linear activation curve
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Linear Activation')
plt.title('Linear Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.axhline(0, color='black', linewidth=0.5)  #value of Y
plt.axvline(0, color='black', linewidth=0.5)  #value of x
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

__3. Tanh Function(-1 to +1)__

- h stands for hyperboilc
- modified version of sigmoid
  

```python
#method1
def tanh_activation(x):
    return np.tanh(x)
print(tanh_activation(-1.8))

#method2
def tanh_activation(x):
    return ((np.exp(2*x) - 1) / (np.exp(2*x) + 1))
print(tanh_activation(-1.8))

#method3
def tanh_custom(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
print(tanh_custom(-1.8))
```

__4. RELU: rectified linear unit (0 to )__

- ReLU(x)=max(0,x)

```python
def relu_activation(x):
    return np.maximum(0, x)
print(relu_activation(9))
print(relu_activation(-5))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def relu_activation(x):
    return np.maximum(0, x)

# Generate x values
x_values = np.linspace(-5, 5, 100)

# Calculate corresponding y values using the ReLU activation function
y_values = relu_activation(x_values)

# Plot the ReLU activation curve
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='ReLU Activation')
plt.title('ReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

__5. Leaky RELU__

__max(alpha*x, x)__
- alpha = 0.01 (smallest & positive , because we don't want it tobe exact 0)

```python
import numpy as np
import matplotlib.pyplot as plt

def leaky_relu_activation(x, alpha=0.01):
    return np.maximum(alpha * x, x)
print(leaky_relu_activation(3))
print(leaky_relu_activation(-3))   #x*alpha
```

```python
import numpy as np
import matplotlib.pyplot as plt

def leaky_relu_activation(x, alpha=0.1):
    return np.maximum(alpha * x, x)

# Generate x values
x_values = np.linspace(-5, 5, 100)

# Calculate corresponding y values using the Leaky ReLU activation function
y_values = leaky_relu_activation(x_values)

# Plot the Leaky ReLU activation curve
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Leaky ReLU Activation')
plt.title('Leaky ReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

__6. Softmax__

- __used for multiclass classification__
- predicting __probablities / confidence score__
- a= 0.6,  b=0.3,  c=0.1.....sum = 1.0
- so here a has most accurate probablity


```python
def softmax(x):
    e_x = np.exp(x)  
    return e_x / e_x.sum(axis=0)   # for 1D always axis = 0

# Generate a vector of three values
x_values = np.array([1.3, 5.1, 2.2, 0.7, 1.1])

# Calculate the softmax probabilities
softmax_values = softmax(x_values)

print(list(zip(x_values, softmax_values)))
print(sum(softmax_values))
```

```python
def softmax(x):
    e_x = np.exp(x - np.max(x))  # Subtracting max(x) for numerical stability
    return e_x / e_x.sum(axis=0)

# Generate a vector of larger values
x_values = np.linspace(-10, 10, 100)

# Calculate the softmax probabilities
softmax_values = softmax(x_values)

# Plot the softmax probabilities
plt.plot(x_values, softmax_values, linestyle='-', color='b')
plt.title('Softmax Function')
plt.xlabel('Input Values')
plt.ylabel('Softmax Probabilities')
plt.show()
```
