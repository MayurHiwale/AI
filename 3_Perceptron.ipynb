```python
class Perceptron:
    def __init__(self, input_size):
        self.weights = np.random.rand(input_size)    #always set random values...for stable output need ranodm.seed
        self.bias = np.random.rand()
        self.learning_rate = 0.1

    def predict(self, inputs):
        weighted_sum = np.dot(inputs, self.weights) + self.bias
        return 1 if weighted_sum >= 0 else 0

    def train(self, training_inputs, labels, epochs):
        for epoch in range(epochs):
            print(f"Epoch {epoch + 1}:")

            for i, (inputs, label) in enumerate(zip(training_inputs, labels)):
                prediction = self.predict(inputs)
                error = label - prediction

                # Displaying current state
                print(f"\tExample {i + 1}:")
                print(f"\t\tInputs: {inputs}")
                print(f"\t\tWeights: {self.weights}")
                print(f"\t\tBias: {self.bias}")
                print(f"\t\tPrediction: {prediction}")
                print(f"\t\tLabel: {label}")
                print(f"\t\tError: {error}")

                # Update weights and bias
                self.weights += self.learning_rate * error * inputs
                self.bias += self.learning_rate * error

                # Display updated state
                print(f"\t\tUpdated Weights: {self.weights}")
                print(f"\t\tUpdated Bias: {self.bias}\n")

# Example usage for logical OR function
if __name__ == "__main__":
    # Define logical OR training data
    training_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
    labels = np.array([0, 1, 1, 1])

    # Create and train the perceptron
    input_size = training_inputs.shape[1]
    perceptron = Perceptron(input_size)
    perceptron.train(training_inputs, labels, epochs=3)  # Adjust epochs for a longer training process
```
